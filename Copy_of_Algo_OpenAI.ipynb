```python
import requests

# Your Marketaux API token
api_token = 'YOUR-KEY'

# API endpoint URL
url = "https://api.marketaux.com/v1/news/all"

# Parameters for the API request
params = {
    'symbols': 'AAPL',
    'filter_entities': 'true',
    'language': 'en',
    'countries': 'us',
    'industries':'Technology',
    'published_on': '2024-04-28',
    'page': '1',
    'api_token': api_token
}

# Making the request to the Marketaux API
response = requests.get(url, params=params)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    print(data)
else:
    print("Failed to retrieve data: ", response.status_code)
```

```python
import json
import re


# Function to remove HTML tags
def remove_html_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

# Extracting the 'data' part of the response
news_items = data.get('data', [])


# Function to print each entity in a structured format
def print_entity(entity):
    print(f"Symbol: {entity['symbol']}")
    print(f"Name: {entity['name']}")
    print(f"Country: {entity['country']}")
    print(f"Type: {entity['type']}")
    print(f"Industry: {entity['industry']}")
    print(f"Match Score: {entity['match_score']}")
    print(f"Sentiment Score: {entity['sentiment_score']}")
    print(f"Highlights: ")
    for highlight in entity.get('highlights', []):
        print(f"\n\n  Highlight:\n {remove_html_tags(highlight.get('highlight', ''))}")
        print(f"  Sentiment: {highlight['sentiment']}")
        print(f"  Highlighted In: {highlight['highlighted_in']}")
    # print("\n")  # New line for separating entities


    # Display each news item in a structured format
for item in news_items:
    print(f"\nTitle: {item['title']}")
    print(f"Description: {item['description']}")
    print(f"Keywords: {item['keywords']}")
    print(f"URL: {item['url']}")
    print(f"Published At: {item['published_at']}")
    print(f"Source: {item['source']}")
    for entity in item.get('entities', []):
        print_entity(entity)
```

```python
import requests

# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol=AAPL&apikey=YOUR-KEY'
r = requests.get(url)
income_statement = r.json()

print(income_statement)
```

```python
for key, value in income_statement.items():
    if key == 'annualReports':
        print(f"{key}:")
        for report in value:
            print("\t{")
            for k, v in report.items():
                print(f"\t\t{k}: {v}")
            print("\t}")
    elif key == 'quarterlyReports':
        print(f"{key}:")
        for report in value:
            print("\t{")
            for k, v in report.items():
                print(f"\t\t{k}: {v}")
            print("\t}")
    else:
        print(f"{key}: {value}")
```

```python
def format_summary(response):
    """
    Formats the macroeconomic summary text into structured sections.

    Parameters:
    response (obj): The response object from OpenAI API.

    Returns:
    None: Prints the formatted summary.
    """
    # Extract text from the response object
    # summary = response.choices[0].text.strip()

    # Split the summary into sections based on '###' (used as markers for headings)
    con=response.content
    sections = con.split('### ')

    for section in sections:
        if section:
            # Split each section into title and content
            parts = section.split('\n', 1)
            title = parts[0].strip()
            content = parts[1].strip() if len(parts) > 1 else ''

            # Print the title and content
            print(f"{title}\n{'-'*len(title)}")
            print(f"{content}\n")
```

```python
def print_output_data_dyn(output_data):
    # Print the Meta Data
    print("Meta Data:")
    for key, value in output_data['Meta Data'].items():
        print(f"  {key}: {value}")

    print("\nTime Series (Daily):")
    # Print the Time Series Data
    for date, data in output_data['Time Series (Daily)'].items():
        print(f"  {date}:")
        for key, value in data.items():
            print(f"    {key}: {value}")
```

```python
from itertools import islice

def format_output_data(output_data, num_dates=20):
    output = "Meta Data:\n"
    for key, value in output_data['Meta Data'].items():
        output += f"  {key}: {value}\n"

    output += "\nTime Series (Daily):\n"
    # Append only the first 'num_dates' of the Time Series Data to the output string
    for date, data in islice(output_data['Time Series (Daily)'].items(), num_dates):
        output += f"  {date}:\n"
        for key, value in data.items():
            output += f"    {key}: {value}\n"

    return output
```
